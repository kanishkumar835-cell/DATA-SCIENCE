```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('C:/Users/kanis/Downloads/Daily_Public_Transport_Passenger_Journeys_by_Service_Type_20250603.csv')
df.head()
```

```python
SORTING
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date").set_index("Date")
df.head()
```

```python
SERVICES THAT WE WANT TO FORECAST
```

```python
services = ["Local Route", "Light Rail", "Peak Service",
"Rapid Route", "School"]
```

```python
forecast_days = 7
all_forecasts = {}

for name in services:
    print("\nForecasting for:", name)
```

```python
history_days_to_show = 60 
forecast_days = 7
```

```python
SCHOOL
```

```python
y = df["School"] 
```

```python
train = y.iloc[:-forecast_days]
last_history = y.iloc[-(history_days_to_show + forecast_days):-forecast_days]
actual_last = y.iloc[-forecast_days:]
```

```python
history_ma7 = last_history.rolling(7).mean()
plt.figure(figsize=(8,3))

last_history.plot(label="daily (last 60 days)", alpha=0.3, color="steelblue")
history_ma7.plot(label="7day moving average", color="blue")
actual_last.plot(label="actual last 7 days", color="black")
forecast.plot(label="forecast next 7 days", color="red")

plt.title("School  recent history and 7day forecast")
plt.xlabel("Date")
plt.ylabel("Passenger journeys")
plt.legend()
plt.tight_layout()
plt.show()
```
