```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('C:/Users/kanis/Downloads/Daily_Public_Transport_Passenger_Journeys_by_Service_Type_20250603.csv')
df.head()
```

```python
SORTING
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date").set_index("Date")
df.head()
```

```python
INSIGHT-1: PEAK SERVICE
PeakService jumps up suddenly at certain times and on certain days.
These short, sharp peaks show that buses or trains are getting very full for a little while, 
so the operator may need to add extra trips or bigger vehicles only in those busy windows, not all day.
```

```python
df["Peak Service"].plot(figsize=(10,3), alpha=0.4, label="daily")
df["Peak Service"].rolling(7).mean().plot(label="7d mean")
plt.legend(); plt.title("Peak Service  daily & rolling"); plt.show()
```

```python
INSIGHT-2: LIGHT RAIL VS BUSES
When more people start using LightRail, fewer people use LocalRoute and RapidRoute buses.
This pattern means many passengers prefer the faster rail option and are switching from bus to LightRail when they can.
```

```python
df[["Light Rail","Local Route","Rapid Route"]].plot(figsize=(10,3))
plt.title("Light Rail vs bus routes"); plt.show()
```

```python
INSIGHT-3: STEADY
Looking at daytoday variation, RapidRoute has the smoothest, so it is the most stable and easiest to predict.
School routes jump up and down the most, so they are the least stable and hardest to plan without considering other factors.
```

```python
df.columns = df.columns.str.strip()
cols = ["Local Route", "Light Rail", "Peak Service",
"Rapid Route", "School", "Other"]
stds = df[cols].std().sort_values()
print(stds)
stds.plot(kind="bar", figsize=(6,3), title="Volatility by service")
plt.ylabel("Std dev of daily journeys")
plt.tight_layout(); plt.show()

```

```python
INSIGHT-4: SCHOOL CALENDER
School ridership clearly follows the academic calendar.
On normal school days it is strong, but during holidays and exam breaks the numbers drop to almost zero,
so planning for these routes must follow school schedules closely.
```
